```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import requests
%matplotlib inline
```

```python
link = []
pages = 1
while pages != 21:
    url = f'https://www.tippsysake.com/collections/all?page={pages}'
    response = requests.get(url)
    # html = response.text
    sake = BeautifulSoup(response.content, 'html.parser')
    for i in sake.find_all('a', {'class': 'product-card__link'}):
        link.append('https://www.tippsysake.com/' + i['href'])
    pages = pages + 1
len(link)
```

```python
name = []
pref = []
brewery = []
infor = []
pair = []
count = 0
for links in link:
    url = links
    response = requests.get(url)
    # html = response.text
    sake = BeautifulSoup(response.content, 'html.parser')
    count+=1
    name.append(sake.find('h1').text.replace('-', ''))
    try:
        brewery.append(sake.find('p', {'class' : 'h2 font-weight-bold mb-1'}).getText())
    except:
        brewery.append(None)
    try:
        pref.append(sake.find('p', {'class' : 'h3 mb-0 font-weight-normal'}).getText())
    except:
        pref.append(None)
    try:
        info = sake.findAll('td', {'class' : 'bg-white'})
        for i in info:
            infor.append(i.getText())
    except:
        info.append(None)
    try:
        pairing = sake.find('ul', {'class' : 'list-inline text-center text-md-left recommended_paring' })
        first = pairing.find('li', {'class' : 'list-inline-item mr-0' }).getText()
        second = pairing.find('li', {'class' : 'list-inline-item mr-0' }).findNext('li').getText()
        pair.append(first + second)
    except:
        pair.append(None)
    print(f'sake {count}')
#     pair.append(x.getText())
```

```python
name
```

```python
classs = []
character = []
alcohol = []
smv = []
acidity = []
list = [infor[i:i+7] for i in range(0, len(infor), 7)]
```

```python
for i in list:
    try:
        classs.append(i[0])
    except:
        classs.append(None)
    try:
        character.append(i[2])
    except:
        character.append(None)
    try:
        alcohol.append(i[3])
    except:
        alcohol.append(None)
    try:
        smv.append(i[5])
    except:
        alcohol.append(None)
    try:
        acidity.append(i[6])
    except:
        alcohol.append(None)
```
