```python

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(sd[lis_1])
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
knn_data = pd.read_csv('./sakefood_cluster.csv')
```

```python
knn_data.dropna(subset = ['Class'], inplace=True)

x = knn_data[lis_w]
y = knn_data['Clusters']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30) # using 30% of the data to test with
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train.values.ravel())
```

```python
y_pred = knn.predict(x_test)
print('accuracy: ', knn.score(x_test, y_test))
```
